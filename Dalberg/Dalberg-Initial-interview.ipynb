# **Data Cleaning**

```python
# Get info on our data
nairobi_df.info()
```

From the onset, it seems we have a total of 942 health facilities in Nairobi. Some columns seem to have a significant number of null values so let's delve into that.

```python
# Percentage of missing values
nairobi_df.isnull().sum()/len(nairobi_df)*100
```

**Dealing with missing values**:

There are columns where all of the rows are null so they should be deleted. We also have quite a number of columns with more that 50% of the data missing. We will drop those as well.

```python
# Drop columns where more that 50% of the data is missing.
nairobi_df = nairobi_df.drop(columns=[col for col in nairobi_df.columns if nairobi_df[col].isnull().mean() > 0.50])
```

# **Data Visualization**

```python
# Number of facilities by owner (percentage)
owner_counts = nairobi_df['Owner'].value_counts()

# Calculate the percentage of facilities owned by each unique owner
total_facilities = len(nairobi_df)
owner_percentages = (owner_counts / total_facilities) * 100

# Create a horizontal bar plot
plt.figure(figsize=(10, 8))
plot = sns.barplot(x=owner_counts.values, y=owner_counts.index, palette='viridis')

# Add annotations with percentages
for i, v in enumerate(owner_counts.values):
    plot.text(v + 10, i, f'{owner_percentages.iloc[i]:.1f}%', va='center', color='black', fontsize=10)

# Add labels and title
plt.xlabel('Number of Facilities')
plt.ylabel('Owner')
plt.title('Number of Facilities by Owner')

# Display the plot
plt.show()
```

```python
# Subset the data for public facilities coz we want to see access to affordable healthcare
govt_facilities_df = nairobi_df[nairobi_df['Owner'] == 'Ministry of Health']
govt_facilities_df.info()
```

We have 56 government facilities which makes up 5.6% of the health facilities in Nairobi County. We will now look at the operational status of these facilities.

```python
# Visialize the operational status of the govt facilities
status_counts = nairobi_df['Operational Status'].value_counts()

# Create a pie chart
plt.figure(figsize=(10, 6))
plt.pie(status_counts, labels=status_counts.index, autopct='%1.1f%%', startangle=140, colors=['#66b3ff','#ff9999','#99ff99'])
plt.title('Distribution of Operational Statuses')

# Add a legend with actual values
plt.legend(title="Status", loc="upper left", labels=[f'{status}: {count}' for status, count in zip(status_counts.index, status_counts.values)])
plt.show()
```

```python
# Count the number of facilities for each type
type_counts = nairobi_df['Type'].value_counts()

# Create a bar plot
plt.figure(figsize=(12, 8))
plot = sns.barplot(x=type_counts.values, y=type_counts.index, palette='viridis')

# Add annotations
for i, value in enumerate(type_counts.values):
    plot.text(value + 0.5, i, f'{value}', va='center', color='black', fontsize=10)

# Add labels and title
plt.xlabel('Number of Facilities')
plt.ylabel('Facility Type')
plt.title('Number of Facilities by Type')

# Display the plot
plt.show()
```

```python
# Count the number of health facilities for each district
district_counts = nairobi_df['District'].value_counts()

# Create a horizontal bar plot
plt.figure(figsize=(12, 10))
plot = sns.barplot(x=district_counts.values, y=district_counts.index, palette='viridis')

# Add annotations
for i, value in enumerate(district_counts.values):
    plot.text(value + 0.5, i, f'{value}', va='center', color='black', fontsize=10)

# Add labels and title
plt.xlabel('Number of Health Facilities')
plt.ylabel('District')
plt.title('Number of Facilities by District')

# Display the plot
plt.show()
```

```python
# Create a pivot table with districts as rows, facility types as columns, and counts as values
pivot_table = nairobi_df.pivot_table(index='District', columns='Type', aggfunc='size', fill_value=0)

# Set up the matplotlib figure
plt.figure(figsize=(14, 10))

# Create the heatmap
sns.heatmap(pivot_table, cmap='viridis', annot=True, fmt='d', linewidths=.5)

# Add labels and title
plt.xlabel('Facility Type')
plt.ylabel('District')
plt.title('Number of Facilities by Type and District')

# Display the plot
plt.show()
```
